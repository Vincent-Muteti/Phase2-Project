```python
import  pandas as pd    
import  numpy  as  np
```

```python
# import the dataset
df = pd.read_csv('Data/tn.movie_budgets.csv')
df
```

```python
# checking null values 
df.isnull().sum()
```

```python
# removing duplicate rows
df = df.drop_duplicates()
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# changing object to int
df['production_budget'] = df['production_budget'].str.replace('$','').str.replace(',','').astype(int)
```

```python
# changing object to int
df['domestic_gross'] = df['domestic_gross'].str.replace('$','').str.replace(',','').astype(int)
```

```python
df.info()
```

```python
# changing dates 
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# changing object to float
df['worldwide_gross'] = df['worldwide_gross'].str.replace('$','').str.replace(',','').astype(float)
```

```python
df.info()
```

```python
df
```

```python
# cleaned data export
df.to_csv('Data/cleaned_movie_budgets.csv', index=False)
```
