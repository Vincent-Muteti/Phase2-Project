### Data Cleaning

```python
#Drop unwanted columns & ignores the error if columns are not found
mg.drop(columns=['studio', 'year'], inplace=True, errors='ignore')


# Keep rows where both domestic and foreign gross are non-missing
mg_revenue = mg.dropna(subset=['domestic_gross', 'foreign_gross'])

#fix data type in foreign_gross
mg['foreign_gross'] = (
    mg['foreign_gross']
    .replace('[\$,]', '', regex=True)   # Remove $ and commas
    .replace('', np.nan)                # Replace empty strings with NaN
    .astype(float)                      # Convert to numbers
)

mg_revenue.shape  # check how many rows are left
```
