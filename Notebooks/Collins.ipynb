```python
#Identify Missing Values
df.isnull().sum()
```

```python
#Checking and dropping of Duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df
```

```python
df.info()
```

```python
df.dtypes
```

```python
object_cols = df.select_dtypes(include='object').columns
print(object_cols)
```

```python
# Remove leading/trailing spaces
df[object_cols] = df[object_cols].apply(lambda x: x.str.strip())

# Replace empty strings with NaN (useful for later numeric conversion)
df[object_cols] = df[object_cols].replace('', pd.NA)
```

```python
num_cols = ['domestic_gross', 'worldwide_gross', 'production_budget']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
df.head()
```

```python
df['domestic_profit'] = df['domestic_gross'] - df['production_budget']
df['worldwide_profit'] = df['worldwide_gross'] - df['production_budget']
df.head()
```
