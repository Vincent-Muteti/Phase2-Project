```python
import  pandas as pd    
import  numpy  as  np
```

```python
# import the dataset
df = pd.read_csv('Data/tn.movie_budgets.csv')
df
```

```python
# checking null values 
df.isnull().sum()
```

```python
# removing duplicate rows
df = df.drop_duplicates()
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# changing object to integer or float
df['production_budget'] = (
    df['production_budget']
    .astype(str)              # convert to string
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)            
)

```

```python
df.info()
```

```python
# changing dates 
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df.info()
```

```python
# domestic profit
df['domestic_profit'] = df['domestic_gross'] - df['production_budget']
df
```

```python
# worldwide profit
df['worldwide_profit'] = df['worldwide_gross'] - df['production_budget']
df
```

```python
df
```

```python
# cleaned data export
df.to_csv('Data/cleaned_movie_budgets.csv', index=False)
```

## Visualizations


### Histoplot of movies vs worldwide gross

```python
# visualizing the cleaned data
import matplotlib.pyplot as plt
import seaborn as sns
# histoplot of movies vs worldwide gross  
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='worldwide_gross', bins=30, kde=True)
plt.title('Distribution of Worldwide Gross')
plt.show()
```

### Top 10 movies per worldwide profit

```python
# plotting a bar plot of movie with highest worldwide profit in descending order
top_movies = df.nlargest(10, 'worldwide_profit')
plt.figure(figsize=(12,6))
sns.barplot(data=top_movies, x='title', y='worldwide_profit')
plt.xticks(rotation=45)
plt.title('Top 10 Movies by Worldwide Profit')
plt.show()
```

### Top 10 movies vs domestic profit

```python
# top 10 movies per domestic profit
top_domestic_movies = df.nlargest(10, 'domestic_profit')
plt.figure(figsize=(12,6))
sns.barplot(data=top_domestic_movies, x='title', y='domestic_profit')
plt.xticks(rotation=45)
plt.title('Top 10 Movies by Domestic Profit')
plt.show()
```
